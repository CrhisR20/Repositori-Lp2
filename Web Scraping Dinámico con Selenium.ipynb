### Armaremos funciones para obtener los datos de las escalas y las tarifas. 

```python
# Esta funcion obtiene los precios de un vuelo en particular

def obtener_precios(vuelo):
    tarifas = vuelo.find_elements_by_xpath('.//ol[@aria-label="Tarifas disponibles."]/li[@tabindex="0"]')
    precios = []
    for tarifa in tarifas:
        nombre = tarifa.find_element_by_xpath('.//span[@class="sc-fElddq bxmgHE"]').text
        moneda = tarifa.find_element_by_xpath('.//span[@class="sc-euofZA GCYF displayAmount"]/span[@class="sc-dqvjwr emWoWC currency"]').text
        valor = tarifa.find_element_by_xpath('.//span[@class="sc-euofZA GCYF displayAmount"]/span[@class="sc-dqvjwr emWoWC"]').text 
        dict_tarifa={nombre:{'moneda':moneda, 'valor':valor}}
        precios.append(dict_tarifa)
        #print(dict_tarifa)
    return precios
```

```python
#Esta funcion obtiene algunos de las escalas de un vuelo en particular

def obtener_datos_escalas(vuelo):
    segmentos = vuelo.find_elements_by_xpath('//section[@data-test="section-info-leg"]')
    info_escalas = []
    for segmento in segmentos:
        # Origen
        origen = segmento.find_element_by_xpath('.//div[@class="sc-hJfILt kKxvzl"]/div[@class="iataCode"]/span').text
        # Aeropuerto
        aeropuerto = segmento.find_element_by_xpath('.//div[@class="sc-hJfILt kKxvzl"]/span[@class="ariport-name"]').text
        # Hora de salida
        hora_salida = segmento.find_element_by_xpath('.//div[@class="sc-hJfILt kKxvzl"]/div[@class="iataCode"]/span[@class="time"]').text
        # Duracion
        duracion_vuelo = segmento.find_element_by_xpath('.//div[@class="sc-cnJOrV gCWXgl"]/span[@class="time"]').text
        
        # Armamos un diccionario para almacenar los datos
        data_dict={
            'origen':origen,
            'aeropuerto':aeropuerto,
            'hora de salida':hora_salida,
            'duracion de vuelo':duracion_vuelo}
        info_escalas.append(data_dict)
        
    return info_escalas
        
```

```python
import time
```

### Construyendo una funcion para unificar el scraper

```python
def obtener_info(driver):
    vuelos = driver.find_elements_by_xpath('//li[@class="sc-sVRsr jVBmqg"]')
    print(f'Se encontraron {len(vuelos)} vuelos.')
    print('Iniciando scraping...')
    info = []
    for vuelo in vuelos:
    
        # Clickeamos el botón de escalas para ver los detalles
        vuelo.find_element_by_xpath('.//a[@data-reference="modal-air-offers"]').click()
        time.sleep(2)
        escalas = obtener_datos_escalas(vuelo)
        # Cerramos el pop-up con los detalles
        driver.find_element_by_xpath('//div[@role="dialog"]/button[@type="button"]').click()
        
        # Clickeamos el vuelo para ver los precios
        vuelo.click()
        precios = obtener_precios(vuelo)
        # Cerramos los precios del vuelo
        driver.find_element_by_xpath('//div[@class="sc-kSpkgE cddBCQ"]/button[@type="button"]').click()
        
        info.append({'precios':precios, 'escalas': escalas})
        
    return info
```
